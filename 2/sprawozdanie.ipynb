1. $x\in<-1,1>$  
2. $x\in<-5,5>$  
3. $x\in<0,5>$
  
***Wskazówki*** 
Jako argumentu funkcji można użyć numpy array.  

Do wizualizacji wyników w Pythonie używa się pakietu [Matplotlib](https://matplotlib.org/). Działanie tego pakietu prezentuje przykład:

```python
x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
plt.plot(x, y, label ='sin(x)')
plt.show()
```

Wspierając się dokumentacją [Matplotlib](https://matplotlib.org/) dodaj do wykresu etykiety osi, tytuł, grida i legendę.

```python
# Definiuje funkcje
def f(x):
    return (x*x*x)-(3*x)


#ad1 Wykres do przedzialu 1
ad1 = np.linspace(-1,1)
plt.plot(ad1, f(ad1), label = 'x')
plt.suptitle('x € <-1, 1>')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

#ad2 Wykres do przedzialu 2
ad2 = np.linspace(-5,5)
plt.plot(ad2, f(ad2), label = 'x')
plt.suptitle('x € <-5, 5>')
plt.grid(color='g', linestyle='-', linewidth=0.3)
plt.legend()
plt.show()

#ad3 Wykres do przedzialu 3
ad3 = np.linspace(0,5)
plt.plot(ad3, f(ad3), label = 'x')
plt.grid(color='r', linestyle='-.', linewidth=0.5)
plt.suptitle('x € <0, 5>')
plt.legend()
plt.show()
```

***Zadanie 3.*** 
Oblicz ilość ciepła, które wydzieli się podczas hamowania jeżeli opisuje je zależność:  
  
$Q = \cfrac{mv^2}{2}$  
  
jeżeli $m = 2500g$, $v = 60km/h$, wypisz wynik oraz podaj w kilokaloriach i dżulach. Wykreśl wykresy które zobrazują zmianę ilości ciepła w procesie hamowania dla ciała o masie $3000g$ i prędkości $v\in<200,0>km/h$, przedstawiając ciepło hamowania na pierwszym w skali liniowej na drugim w skali logarytmicznej.  

***Wsakzówka:*** Do stworzenia wykresów z osią logarytmiczną skorzystaj z funkcji: [semilogy](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.semilogy.html), [xlim](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim#matplotlib.pyplot.xlim) i [figure](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html?highlight=figure#matplotlib.pyplot.figure) z pakietu *Matplotlib*. 

Rozwiązanie:

```python
# Definiuje funkcje dla podanego wzoru na cieplo
def q(m, v):
    return (m*(v*v))/2


qj = q(2500/1000, 60*(10/36))

# Konwertuje otrzymany wynik na wynik w kilokaloriach
qkcal = (0.238846*qj)/1000 # 0.238846 - ilość kalorii w 1 J
print("Wynik w dżulach: Q = {:.2f} J".format(qj))
print("Wynik w kilokaloriach: Q = {:.3f} kcal".format(qkcal))


# Wykresy dla m = 3000g
qj2 = q(3000/1000, (np.linspace(200, 0))*10/36)

# W skali liniowej
plt.plot(np.linspace(0, 200), qj2)
plt.grid(color='black', linestyle='-.', linewidth=0.5)
plt.xlim(200, 0)
plt.show()

# W skali logarytmicznej
plt.semilogy(np.linspace(0, 200), qj2)
plt.grid(color='black', linestyle='-.', linewidth=0.5)
plt.xlim(200, 0)
plt.show()
```

Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x,y,xlabel,ylabel,title,label):
    if x.shape != y.shape or  min(x.shape)==0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig

x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x,y,'x','y','Funkcja sin', 'sin(x)')
```

Za pomocą funkcji *compare\_plot* rozwiąż graficznie równanie: $f(x) = g(x)$, gdzie:  
$f(x) = x+2$  
$g(x) = x^2 -2sin(x) + 3$

Rozwiązanie: 

```python
# Definiuje funkcje dla danych f(x) i g(x)
def ad4f(x):
    return x+2

def ad4g(x):
    return (x*x) - (2*np.sin(x)) + 3

# Przypisuje do zmiennych wartosci funkcji
x1 = np.linspace(-10, 10)
y1 = ad4f(x1)
x2 = np.linspace(-10, 10)
y2 = ad4g(x2)

# Korzystam z uprzednio napisanej funkcji compare_plot z pliku main.py
main.compare_plot(x1, y1, x2, y2, 'x', 'y', 'Zadanie 4', 'f(x) = x + 2', 'g(x) = x^2 - 2sin(x) + 3')
```

***Zadanie 6.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

Rozwiązanie:

```python
# Definiuje funkcję spirali z pomoca rownan parametrycznych z wikipedii
def logs(w, s):
    space = np.linspace(0, 20)
    x = w * np.exp(s * space) * np.cos(space)
    y = w * np.exp(s * space) * np.sin(space)
    return x, y
# Generuje spirale
x1, y1 = logs(1, 0.1)
x2, y2 = logs(1, 0.3)

# Korzystam z uprzednio napisanej funkcji parallel_plot z pliku main.py

main.parallel_plot(x1, y1, x2, y2, 'x', 'y', 'x', 'y', 'Spirala logarytmiczna', 'v');
```
