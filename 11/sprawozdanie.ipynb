Ad. 1
Całkowanie podanej funkcji na dowolnym przedziale nie jest możliwe, ponieważ w zadanej funkcji
występują punkty nieciągłości

```python
import numpy as np
from scipy import integrate

fx = lambda x: np.log10(1 + np.tan(x))

# Długości wektorów przyjmuję pod metodę Romberga

# 2049
x1 = np.linspace(0, np.pi/4, 2049)
dx1 = x1[1]
y1 = fx(x1)

print('Całka numeryczna dla wektora 2049:\n')

i_x1_t1 = integrate.trapz(y1)
i_x1_t2 = integrate.trapz(y1, dx=dx1)
i_x1_t3 = integrate.trapz(y1, dx=dx1*1000)

i_x1_s1 = integrate.simps(y1)
i_x1_s2 = integrate.simps(y1, dx=dx1)
i_x1_s3 = integrate.simps(y1, dx=dx1*1000)

i_x1_r1 = integrate.romb(y1, 1)
i_x1_r2 = integrate.romb(y1, dx1)
i_x1_r3 = integrate.romb(y1, dx1*1000)

print('Dla dx = {}:\n'.format(1))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(1, i_x1_t1))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(1, i_x1_s1))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(1, i_x1_r1))

print('Dla dx = {}:\n'.format(dx1))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(dx1, i_x1_t2))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(dx1, i_x1_s2))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(dx1, i_x1_r2))

print('Dla dx = {}:\n'.format(dx1*1000))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(dx1*1000, i_x1_t3))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(dx1*1000, i_x1_s3))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(dx1*1000, i_x1_r3))



# 1025
x2 = np.linspace(0, np.pi/4, 1025)
dx2 = x2[1]
y2 = fx(x2)

print('Całka numeryczna dla wektora 1025:\n')

i_x2_t1 = integrate.trapz(y2)
i_x2_t2 = integrate.trapz(y2, dx=dx2)
i_x2_t3 = integrate.trapz(y2, dx=dx2*1000)

i_x2_s1 = integrate.simps(y2)
i_x2_s2 = integrate.simps(y2, dx=dx2)
i_x2_s3 = integrate.simps(y2, dx=dx2*1000)

i_x2_r1 = integrate.romb(y2, 1)
i_x2_r2 = integrate.romb(y2, dx2)
i_x2_r3 = integrate.romb(y2, dx2*1000)

print('Dla dx = {}:\n'.format(1))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(1, i_x2_t1))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(1, i_x2_s1))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(1, i_x2_r1))

print('Dla dx = {}:\n'.format(dx2))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(dx2, i_x2_t2))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(dx2, i_x2_s2))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(dx2, i_x2_r2))

print('Dla dx = {}:\n'.format(dx2*1000))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(dx2*1000, i_x2_t3))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(dx2*1000, i_x2_s3))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(dx2*1000, i_x2_r3))



# 65
x3 = np.linspace(0, np.pi/4, 65)
dx3 = x3[1]
y3 = fx(x3)

print('Całka numeryczna dla wektora 65:\n')

i_x3_t1 = integrate.trapz(y3)
i_x3_t2 = integrate.trapz(y3, dx=dx3)
i_x3_t3 = integrate.trapz(y3, dx=dx3*1000)

i_x3_s1 = integrate.simps(y3)
i_x3_s2 = integrate.simps(y3, dx=dx3)
i_x3_s3 = integrate.simps(y3, dx=dx3*1000)

i_x3_r1 = integrate.romb(y3, 1)
i_x3_r2 = integrate.romb(y3, dx3)
i_x3_r3 = integrate.romb(y3, dx3*1000)

print('Dla dx = {}:\n'.format(1))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(1, i_x3_t1))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(1, i_x3_s1))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(1, i_x3_r1))

print('Dla dx = {}:\n'.format(dx3))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(dx3, i_x3_t2))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(dx3, i_x3_s2))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(dx3, i_x3_r2))

print('Dla dx = {}:\n'.format(dx3*1000))
print('Wynik całkowania metodą Trapezów dla dx = {} wynosi: \n{}\n'.format(dx3*1000, i_x3_t3))
print('Wynik całkowania metodą Simpsona dla dx = {} wynosi: \n{}\n'.format(dx3*1000, i_x3_s3))
print('Wynik całkowania metodą Romberga dla dx = {} wynosi: \n{}\n'.format(dx3*1000, i_x3_r3))
```

Ad. 2

Wyniki całkowania zależnie od metody różnią się w nikłym stopniu od siebie (15/16 miejsce po przecinku). 
Długość wektora y ma wpływ na wyniki całkowania poszczególnymi metodami, szczególnie
jest to widoczne przy domyślnym parametrze dx = 1. Parametr dx ma wpływ liniowy na wyniki całkowania.

Ad.3

```python
i_quad = integrate.quad(fx, 0, np.pi/4)
print('Wynik całkowania funkcją Quad = {}\n'.format(i_quad[0]))

i_romberg = integrate.romberg(fx, 0, np.pi/4)
print('Wynik całkowania funkcją Romberg = {}\n'.format(i_romberg))

i_quadrature = integrate.quadrature(fx, 0, np.pi/4)
print('Wynik całkowania funkcją Quadrature = {}\n\n'.format(i_quadrature[0]))

t1 = %timeit -r 100 -n 100 -o integrate.quad(fx, 0, np.pi/4)
t2 = %timeit -r 100 -n 100 -o integrate.romberg(fx, 0, np.pi/4)
t3 = %timeit -r 100 -n 100 -o integrate.quadrature(fx, 0, np.pi/4)

t4 = %timeit -r 100 -n 100 -o integrate.trapz(y1, dx=dx1)
t5 = %timeit -r 100 -n 100 -o integrate.simps(y1, dx=dx1)
t6 = %timeit -r 100 -n 100 -o integrate.romb(y1, dx1)
```

Rozwiązanie całki powyższymi metodami jest zbliżone do tego w podpunkcie 2 i 
różni się dopiero po ok. 15 miejscu po przecinku.

```python
import matplotlib.pyplot as plt

t = [t1.average, t2.average, t3.average, t4.average, t5.average, t6.average]
t_x = ['Quad', 'Romberg', 'Quadrature', 'Trapz', 'Simps', 'Romb']

plt.figure()
plt.plot(t_x, t)
plt.grid()
plt.title('Średni czas wykonania całkowania, zależny od metody')
plt.ylabel('Czas')
```

Metody Trapz oraz Simps są najbardziej wydajne jeśli chodzi o czas i wynik całkowania. Metodę Romberg 
można postawić na następnym miejscu. Metoda Romb jest najmniej wydajna.
